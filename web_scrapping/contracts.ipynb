```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta

tm_fbref = pd.read_excel("tm_fbref.xlsx")
contracts = pd.read_excel("contracts.xlsx")
```

```python
# spain["contract_until"]
coincidences = []

ct_urls = list(contracts["url"])
tm_urls = list(tm_fbref["UrlTmarkt"])

for url_s in ct_urls:
    b = False
    for url_t in tm_urls:
        if url_t.split('/')[6] == url_s.split('/')[6]:
            coincidences.append(url_s)
            b = True
            break
    if b: continue
#     print(url_s)
```

```python
contracts.head()
```

```python
data = {  
    'url': [],  
    'expiration': []  
}    

for row in range(contracts.shape[0]):
    url = contracts.iloc[row]["url"]
    exp = contracts.iloc[row]["contract_until"]
    
    if url in coincidences and exp == exp:
        data["url"].append(url)
        data["expiration"].append(exp)

df = pd.DataFrame(data)
```

```python
# Fechas de ejemplo
fecha_inicio = datetime(2022, 5, 15)
fecha_fin = datetime(2024, 1, 10)

# Calcular la diferencia en meses
diferencia_meses = relativedelta(fecha_fin, fecha_inicio).years * 12 + relativedelta(fecha_fin, fecha_inicio).months

print(diferencia_meses)
```

```python
expirations = list(df["expiration"])
months = []

fecha_inicio = datetime(2025, 1, 30)

for exp in expirations:
    items = exp.split("-")
    
    year = int(items[0])
    month = int(items[1])
    day = int(items[2])
    
    fecha_fin = datetime(year, month, day)
    
    diferencia_meses = relativedelta(fecha_fin, fecha_inicio).years * 12 + relativedelta(fecha_fin, fecha_inicio).months
    months.append(diferencia_meses)
```

```python
df["months"] = months
```

```python
df.to_excel('contracts.xlsx', index=False)  
```
